```python
#open and read the file for tuesday, set data types correctly
tuesday = pd.read_csv('/home/buli/Documents/Spiced/convex_capers_student_code/data/marketplace/tuesday.csv',sep = ';')
tuesday['timestamp'] = pd.to_datetime(tuesday['timestamp'],infer_datetime_format=True)
```

```python
# a sample of 100 rows to perform calculations
sample = tuesday.iloc[0:100]
sample
```

```python
# see location transition behaviour
sample.groupby(['location']).count()
```

```python
# in order for the location to pass consequently according to time, we need to sort according to ucstomer number
#  (order does not matter), and then by time, to make sure we have the correct order
sample.sort_values(['customer_no','timestamp'],inplace=True)
for customer in sample['customer_no']:
    sample['first_location'] = sample['location'].shift(1)
    sample['first_location'].fillna(value=0,inplace = True)

sample
```

```python
# probability of visiting one location, given that a customer was in a previous location

probability = pd.crosstab(sample['first_location'], sample['location'], normalize=0)
probability
```
